```python
import torch
tensor_A = torch.rand(7,7)
tensor_A
```

```python
tensor_B = torch.rand(1,7)
tensor_B
```

```python
tensor_C = torch.matmul(tensor_A, tensor_B.T)
tensor_C
```

```python
torch.manual_seed(2)
X=torch.rand(size=(7,7))
Y=torch.rand(size=(1,7))
Z=torch.matmul(X,Y.T)
Z, Z.shape
```

```python
max=torch.max(Z)
min=torch.min(Z)
max, min
```

```python
arg_max=torch.argmax(Z)
arg_min=torch.argmin(Z)
arg_max, arg_min
```
