1. Tthe **inner dimensions** must match:
* `(3,2) @ (3,2)` won't work
* `(3,2) @ (2,3)` will work
* `(2,3) @ (3,2)` will work
2. The resulting matrix has the shape of the **outer dimensions**
* `(2,3) @ (3,2)` -> `(2,2)`
* `(3,2) @ (2,3)` -> `(3,3)`

## Reshaping, stacking ,squeezing and unsqueezing tensors

* Reshaping - reshqpes an input tensor to a defined shape
* View - Return a view of an input tensor of certain shape but keep the same memory as the original tensor
^ Stacking - combine multiple tensors on top of each other (vstack) or side by side (hstack)
https://docs.pytorch.org/docs/2.11/generated/torch.stack.html
https://docs.pytorch.org/docs/2.12/generated/torch.vstack.html
https://docs.pytorch.org/docs/2.12/generated/torch.hstack.html
* Squeeze - removes all `1` dimensions from a tensor
* Unsqueeze - add a `1` dimension to a target tensor
* Permute - Return a view of the input with dimensions permuted (swapped in a certain way



```python
import torch
x = torch.arange(1., 10.)
x, x.shape
```

```python
x1 = x.reshape(10,1)
x1, x1.shape
```

```python
x2 = x1.reshape(5,2)
x2, x2.shape
```

```python
z = x1.view(1,10)
z, z.shape
```

```python
x
```

```python
x_reshaped = x.reshape(9,1)
x_reshaped, x_reshaped.shape
```

```python
x_reshaped = x.reshape(3,3)
x_reshaped, x_reshaped.shape
```

```python
# Change the view
z = x.view(1,9)
z, z.shape
```

```python
# Changing z changes x (because a view of a tensor shapes the same memory as the original input)
z[:, 0] = 5
z, x
```

```python
# Stack tensors on top of each other
x_stacked = torch.stack([x, x, x, x], dim=0)
x_stacked
```

```python
# torch.squeeze() - remove all single dimentions from a target tensor
x_reshaped = x.reshape(1,9)
x_reshaped, x_reshaped.shape
```

```python
x_squeeze = x_reshaped.squeeze()
x_squeeze, x_squeeze.shape
```

```python
# torch.squeeae() - removes all single dimentions from a target tensor
# https://docs.pytorch.org/docs/2.12/generated/torch.squeeze.html
print(f"Previous tensor: {x_reshaped}")
print(f"Previous shape: {x_reshaped.shape}")

# Remove extra dimentions from x_reshaped
x_squeezed = x_reshaped.squeeze()
print(f"\nNew tensor: {x_squeezed}")
print(f"New shape: {x_squeezed.shape}")

```

```python
# torch.unsqueese() - adds a single dimention to a target tensor at specific dim (dimention)
# https://docs.pytorch.org/docs/2.12/generated/torch.unsqueeze.html
print(f"Previous tensor: {x_squeezed}")
print(f"Previous shape: {x_squeezed.shape}")

# Remove extra dimentions from x_reshaped
# https://docs.pytorch.org/docs/2.12/generated/torch.reshape.html
x_unsqueezed = x_squeezed.unsqueeze(dim=0)
print(f"\nNew tensor: {x_unsqueezed}")
print(f"New shape: {x_unsqueezed.shape}")
#
```

```python
# torch.permute - rearrages the dimentions of a target tensor in a specified order
# https://docs.pytorch.org/docs/2.12/generated/torch.permute.html
x_original = torch.rand(size=(224,224,3)) # [height, width, colour_channels]

# Permute the original tensor to rearrange the axis (or dim) order
x_permuted = x_original.permute(2,0,1) # shifts axis 0->1, 1->2, 2->0

print(f"Previous shape: {x_original.shape}")
print(f"New shape: {x_permuted.shape}") # [colour_channels, height, width]
```

## Indexing (selecting data from tensors)
### Indexing with PyTorch is similar to indexing with NumPy


```python
import torch
x = torch.arange(1,10).reshape(1,3,3)
x, x.shape
```

```python
# Let's index on our new tensor
x[0]
```

```python
# Let's index on the middle bracket (dim=1)
x[0][0]
```

```python
x[0][0][0]
```

```python
# You can also use ":" to select "all" of a target dimension
x[:, 0]
```

```python
# Get all values of 9th and 1st dimensions but only index 1 of 2nd dimension
x[:, : , 1]
```

```python
x[:, 1, 1]
```

```python
# Get index 0 of 0th and 1st dimension and all values of 2nd dimension
x[0,0,:]
```

```python
# Index on x to return 9
x[0][2][2]

# Index on x to return 3,6,9
x[:, :, 2]
```

## Reproducibility

```python
random_tensor_A = torch.rand(3,4)
random_tensor_B = torch.rand(3,4)

print(random_tensor_A)
print(random_tensor_B)
print(random_tensor_A == random_tensor_B)
```

```python
RANDON_SEED = 42
torch.manual_seed(RANDON_SEED)
random_tensor_A = torch.rand(3,4)

torch.manual_seed(RANDON_SEED)
random_tensor_B = torch.rand(3,4)

print(random_tensor_A)
print(random_tensor_B)
print(random_tensor_A == random_tensor_B)
```
